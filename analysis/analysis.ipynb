```python
import pandas as pd
import plotly.express as px
from matplotlib import pyplot as plt
import numpy as np
from scipy.signal import lfilter
```

```python
#opening the files
hashmap = pd.read_csv("addPatients-HM1.csv")
bst = pd.read_csv("addPatients-BST1.csv")
avl = pd.read_csv("addPatients-AVL1.csv")

mean = bst["t"].mean()

std = bst["t"].std()

bst = bst[(bst["t"] <= mean + (1.5*std) )]


mean = hashmap["t"].mean()

std = hashmap["t"].std()

hashmap = hashmap[(hashmap["t"] <= mean + (1.5*std) )]


mean = avl["t"].mean()

std = avl["t"].std()

avl = avl[(avl["t"] <= mean + (1.5*std) )]
```

```python


df = pd.DataFrame()
df['n'] = hashmap['n']
df['t'] = pd.Series([135 for i in range(100000)])

n = 15  # the larger n is, the smoother curve will be
b = [1.0 / n] * n
a = 1

hashmap1 = hashmap

# hashmap1["t"] = lfilter(b, a, hashmap["t"])

hashmap1 = hashmap1[hashmap1['n'] >= 1500]

# hashmap['t'] = lfilter(b, a, hashmap["t"])
fig = px.line(hashmap1, x = 'n', y = 't', title='Graph that shows how the the execution time of adding new patients with respect to the size of the hashmap')

fig.add_scatter(x=hashmap1["n"], y=hashmap1["t"], name="Patient Insertion Time" )
fig.add_scatter(x=df['n'], y=df['t'], name="y=c" )

fig.update_yaxes(title_text="Insetion time (ms)")
fig.update_xaxes(title_text="Hashmap size")
fig.show()
```

```python


df = pd.DataFrame()
df['n'] = bst['n']
df['t'] = 690 * (0.1*bst['n']).apply(np.log)

n = 20  # the larger n is, the smoother curve will be
b = [1.0 / n] * n
a = 1

bst1 = bst

# bst1["t"] = lfilter(b, a, bst["t"])

# bst['t'] = lfilter(b, a, bst["t"])
fig = px.line(bst1, x = 'n', y = 't', title='Graph that shows how the the execution time of adding new patients with respect to the size of the Binary Search Tree')

fig.add_scatter(x=bst1['n'], y=bst1['t'], name="Patient Insertion Time" )
fig.add_scatter(x=df['n'], y=df['t'], name="y=c*log(x)" )

fig.update_yaxes(title_text="Insertion Time (ms)")
fig.update_xaxes(title_text="BST Size")
fig.show()
```

```python


df = pd.DataFrame()
df['n'] = avl['n']
df['t'] = 37 * (0.04 * avl['n']).apply(np.log)

n = 20  # the larger n is, the smoother curve will be
b = [1.0 / n] * n
a = 1

avl1 = avl

avl1["t"] = lfilter(b, a, avl["t"])

# avl['t'] = lfilter(b, a, avl["t"])
fig = px.line(avl1, x = 'n', y = 't', title='Graph that shows how the the execution time of adding new patients with respect to the size of the AVL Tree')

fig.add_scatter(x=avl1['n'], y=avl1['t'], name="Patient Insertion Time" )
fig.add_scatter(x=df['n'], y=df['t'], name="y=c*log(x)" )

fig.update_yaxes(title_text="Insertion Time (ms)")
fig.update_xaxes(title_text="AVL Tree Size")
fig.show()
```

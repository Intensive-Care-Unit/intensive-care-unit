```python
import pandas as pd
import plotly.express as px
from matplotlib import pyplot as plt
import numpy as np
from scipy.signal import lfilter
```

```python
#opening the files
hashmap = pd.read_csv("addPatients-HM1.csv")
bst = pd.read_csv("addPatients-BST1.csv")
avl = pd.read_csv("addPatients-AVL1.csv")

mean = bst["t"].mean()

std = bst["t"].std()

bst = bst[(bst["t"] <= mean + (1.5*std) )]


mean = hashmap["t"].mean()

std = hashmap["t"].std()

hashmap = hashmap[(hashmap["t"] <= mean + (1.5*std) )]


mean = avl["t"].mean()

std = avl["t"].std()

avl = avl[(avl["t"] <= mean + (1.5*std) )]
```

```python


df = pd.DataFrame()
df['n'] = hashmap['n']
df['t'] = pd.Series([135 for i in range(100000)])

n = 15  # the larger n is, the smoother curve will be
b = [1.0 / n] * n
a = 1

hashmap1 = hashmap

hashmap1["t"] = lfilter(b, a, hashmap["t"])

# hashmap1 = hashmap1[hashmap1['n'] >= 1500]

# hashmap['t'] = lfilter(b, a, hashmap["t"])
fig = px.line(hashmap1, x = 'n', y = 't', title='Graph that shows how the the runtime of adding new patients changes against the size of the hashmap')

fig.add_scatter(x=df['n'], y=df['t'], name="Test" )

fig.update_yaxes(title_text="time (ms)")
fig.update_xaxes(title_text="hashmap size")
fig.show()
```
